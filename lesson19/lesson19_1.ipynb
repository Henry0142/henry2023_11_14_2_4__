```python
import numpy as np
import pandas as pd
scores = np.random.randint(50,101,size=(50,5))
scores_pd = pd.DataFrame(scores,
             columns=['國文', '英文', '數學', '地理', '歷史'],
             index = range(1,51)
             )
scores_pd
```

```python
scores_pd.index.name="學號"
scores_pd.columns.name="科目"
scores_pd
```

```python
display(scores_pd['國文'][:10])
scores_pd['數學'][:10]
scores_pd.info()
```

```python
scores_pd['國文'][:10] + scores_pd['數學'][:10]
```

```python
scores_pd['國文'] + scores_pd['英文'] + scores_pd['地理'] + scores_pd['歷史'] + scores_pd['數學']
```

```python
#統計的實體方法,sum()
scores_pd['總分'] = scores_pd[['國文','英文','數學','地理','歷史']].sum(axis=1)
scores_pd
scores_pd['平均'] = scores_pd[['國文','英文','數學','地理','歷史']].mean(axis=1)
scores_pd
```

```python
scores_pd['排名'] = scores_pd['總分'].rank(method='min',ascending=False)
scores_pd
```

```python
scores_pd.sort_values(by='排名')
```

```python
def format60(v:int):
    if v >= 60:
        return '及格'
    else:
        return '不及格'

scores_pd[['國文','英文','數學','地理','歷史']].style.format(format60)
```

```python
scores_pd.style\
.highlight_between(left=0,
                   right=59,
                   color='red',
                   subset=['國文','英文','數學','地理','歷史'],
                   props='color:white;background-color:red')\
.format(precision=0,subset=['平均','排名']).to_excel('scores.xlsx')
```
